```python
nx=5
xo=2.0
X=[]
Y=[]
u=[]
v=[]

t=linspace(-4,4,nx)

#funcion a graficar

def f(x,y):  
    if(x==0 and y==0):
        return 0,0
    if(x==xo and y==0):
        return 0,0
    ax=x/(x**2+y**2)**(3/2)-(x-xo)/((x-xo)**2+y**2)**(3/2)
    ay=y/(x**2+y**2)**(3/2)-(y)/((x-xo)**2+y**2)**(3/2)
    
    return ax,ay


#haciendo el grid    
for i in range(-nx,nx):
    
    for j in range(-nx,nx):
        X.append(i*1.0)
        Y.append(j*1.0)
    

    
```

```python
print type(X[0])
```

```python
for i in range(len(X)):
    Val=f(X[i],Y[i])
 
    u.append(Val[0])
    v.append(Val[1])

    
print size(u),size(v)
```
