```python
k=linspace(0.1,10,100)
tet=linspace(-pi,pi,100)
f=zeros(len(k))
a=1.0
b=1.0
```

```python
for i in range(len(k)):
        f[i]=a*(1-(4*(k[i]/b)**2 +1)**(-3)) *k[i]**(-2)
        
```

```python
plot(k,f,label="$\sigma$")
title("$ Cross\ section\ $",size=15)
xlabel("$ k$",size=15)
ylabel("$\sigma$",size=15)
legend()
savefig("cross_section")
```

```python
M=zeros((100,100))
A=1
B=1
```

```python
for i in range(len(k)):
    for j in range(len(tet)):
        M[i][j]=A*((2*k[i]*sin(tet[j]/2.0))**2+B**2)**(-4)
```

```python
imshow(M,extent=(-pi,pi,10,0),aspect=0.5)
xlabel("$ \Theta $",size=15)
ylabel("$ \kappa $",size=15)
title("$ d\sigma/d\Omega $",size=15)
```

```python
for i in range(len(k)):
    for j in range(len(tet)):
        M[len(k)-i-1][len(tet)-j-1]=A*((2*k[i]*sin(tet[j]/2.0))**2+B**2)**(-4)
```

```python
imshow(M,extent=(-pi,pi,10,0),aspect=0.5)

xlabel("$ \Theta $",size=15)
ylabel("$ \kappa $",size=15)
title("$ d\sigma/d\Omega $",size=15)
savefig("dif_sigma")
```
