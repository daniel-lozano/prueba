```python
nx=50
xo=1.0
X=[]
Y=[]


t=linspace(-80,80,nx)

#funcion a graficar

def f(x,y,a):
    #primera carga

    if(x==0 and y==0):
        return 0,0
    #segunda carga 
    
    if(x==xo and y==0):
        return 0,0
    #componentes en X y Y
    ax=a*x/(x**2+y**2)**(3/2)-a*(x-xo)/((x-xo)**2+y**2)**(3/2)
    ay=a*y/(x**2+y**2)**(3/2)-a*(y)/((x-xo)**2+y**2)**(3/2)
    
    return ax,ay,ay/ax


#haciendo el grid    
for i in range(len(t)):
    
    for j in range(len(t)):
        X.append(t[i])
        Y.append(t[j])
    

    
    
```

```python
print size(Y),size(X)

print Y[nx],Y[nx-1]

scatter(X[0],Y[0],c="r",s=70)
scatter(X[nx-1],Y[nx-1],c="r",s=70)

#scatter(X[nx+1],Y[nx+1],c="r",s=70)
#print X[nx+1],Y[nx+1]
#scatter(X[nx-1],Y[nx-1],c="r",s=70)
#print X[nx-1],Y[nx-1]
scatter(X,Y)
figsize(5,5)
```

```python
#V=f(3,4,1)
u=[]
v=[]
xo=1
for i in range(len(X)):    
    ang=arctan(f(X[i],Y[i],1000))
    u.append(2*ang[0])
    v.append(2*ang[1])
 
    
    
#for i in range(len(t)):
#     for j in range(len(t)):
#        val=f(X[(i+1)*(nx-1)],Y[i+j],1)
#        u.append(val[0]/abs(val[0]))
#        v.append(val[1]/abs(val[1]))
    
    
    
    
print size(u),size(v)

quiver(X,Y,u,v,color="k",scale=150)
px=[0,xo]
py=[0,0]
scatter(px[0],py[0],c="r",s=80,label="carga Q+")
scatter(px[1],py[1],c="b",s=80,label="carga Q-")
figsize(10,10)

xlim(min(X),max(X))
ylim(min(Y),max(Y))
title("$ Campo\ electrico\ de\ un\ dipolo,\ distancia\ dipolo=\ $ "+str(xo),size=20)
legend()
xlabel(" $ x $", size=25)

ylabel(" $ y $", size=25)
```
