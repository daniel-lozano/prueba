```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=[-11.38,-35.4,1.6,2.7,5.3,-3.8,-3.7,-9.8,2.5,5]
len(X)
```

```python
def func_1(y,X):
    return 1/(y**2+X**2)

def func(y,X):
    
    A=np.ones(len(y))
    
    for j in range(len(y)):
        b=1.0
    
        for i in range(len(X)):
            b=b*(1.0/(y[j]**2+X[i]**2))
            
        A[j]=b
        
    return A
```

```python
N=1000
y=np.linspace(0,100,N)
len(y)
Tot=func(y,X)
plt.plot(y,Tot)
plt.show()
```

```python
F=np.ones(len(y))
Y=np.ones(len(y))
for i in range(2):
    OLD=func_1(y,X[i])
    F=F*OLD
Y=y*F
plt.plot(y,F,label="$ P(y_f|x_1,\dots,x_{10}) $")
plt.plot(y,Y,label="$ y_fP(y_f|x_1,\dots,x_{10}) $")
plt.legend()
plt.show()
```
