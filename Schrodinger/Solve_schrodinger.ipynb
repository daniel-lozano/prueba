```python
steps=200
# atomic units
hbar=1.0
m=1.0
A=1
D=100
W=A/2.0#Well from -W to W
xvec=np.linspace(-A,A,steps) # creates de position vector
h=xvec[1]-xvec[0] #define the step
U=D*(step_func(xvec+W)-step_func(xvec-W))
plt.plot(xvec,U)
plt.ylim(-1,D+10)
plt.show()
plt.close()
```

```python
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)\
        +np.diag(np.ones(steps-1),-1))/(float)(h**2)
Laplacian

# create the Hamiltonian
Hamiltonian=np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U #en la diagonal del Hamiltoniano da el potentencial U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
E,V=diagonalize_hamiltonian(Hamiltonian)
n=0
np.shape(E)
print(E[0],min(E))
#print(E)

for e in range(6):
    Line=np.ones(len(V[:,e]))*0.2*e
    plt.plot(xvec,V[:,e]+0.2*e,label="E="+str(round(E[e],2)))
    plt.plot(xvec,Line,"k--")
    plt.plot()
    
plt.plot(xvec,U)
plt.legend(loc=4)
plt.ylim(-0.05,1.)
plt.xlim(-1,2)
plt.show()
plt.close()
```
