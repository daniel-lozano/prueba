```python
import sys
version=sys.version_info.major
import os
import numpy as np
import scipy as sp
import scipy.linalg as spla
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib as mpl
```

```python
hbar=1
m=1
w=1

def diagonalize_hamiltonian(Hamiltonian):
    return spla.eigh(Hamiltonian)

def step_func(x):
    return 0.5*(1+np.sign(x))

def harmonic(xvec):
    return 0.5*m*(w*xvec)**2

def Potential(r,Vo,a,b):
    return Vo/(1+np.exp((r-b)/a))

def diagonalize_hamiltonian(Hamiltonian):
    return spla.eigh(Hamiltonian)# returns que eigen values and vectors of a hermitian matrix

def Solve(steps,xvec,U,N):
    hbar=1
    m=1
    h=xvec[1]-xvec[0] #define the step
    
    Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)\
        +np.diag(np.ones(steps-1),-1))/(float)(h**2)
    

    # create the Hamiltonian
    Hamiltonian=np.zeros((steps,steps))
    [i,j]=np.indices(Hamiltonian.shape)
    Hamiltonian[i==j]=U #en la diagonal del Hamiltoniano da el potentencial U
    Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
    E,V=diagonalize_hamiltonian(Hamiltonian)
    
    
    for e in range(N):
        Line=np.ones(len(V[:,e]))*E[e]
        plt.plot(xvec,V[:,e]+E[e],label="E="+str(round(E[e],2)))
        plt.plot(xvec,Line,"k--")
        plt.plot()
    
    plt.plot(xvec,U,"k")
    plt.legend(loc=4)
    #plt.ylim(E[0]*(0.9),E[N])
    #plt.xlim(-1,2)
    plt.show()
    plt.close()
```

# $ Infinite\ box\ potential  $

```python
A=2
Num=1
steps=2000
xvec=np.linspace(-A,A,steps)
U=np.zeros(steps)
plt.ylim(0,1.5)
plt.xlim(-A,A)
plt.ylim(0.3,0.35)
Solve(steps,xvec,U,Num)
for i in range(Num):
    print ((i+1)*np.pi/(2*A))**2/2

```

#         $ Rectangular\ Potential\ in\ a\ box\ $

```python
A=10
Num=6
D=100

steps=2000
W=A/2.0
xvec=np.linspace(-A,A,steps)
U=D*(step_func(xvec+W)-step_func(xvec-W))
#plt.xlim(-A,A)
plt.ylim(0,2)
Solve(steps,xvec,U,Num)
for i in range(Num):
    print ((i+1)*np.pi/(2*A))**2/2

```

# $ Harmonic\ oscillator $

```python
Harmonic=harmonic(xvec)
plt.plot(xvec,Harmonic)
plt.show()
```

```python
steps=2000
xvec=np.linspace(-10,10,steps)
Harmonic=harmonic(xvec)
plt.ylim(0,3)
plt.xlim(-4,4)
Solve(steps,xvec,Harmonic,3)


```

```python
Vo=37.0*3.675E4 #MeV
a=0.84*1.9E-5  #fm
b=1.4*1.9E-5 #fm
print(Vo,a,b)
```

```python
steps=3000
xvec=np.linspace(0,2E-4,steps)
U=Potential(xvec,Vo,a,b)
#plt.xlim(0,1.5)
#plt.ylim(5,max(U))
Solve(steps,xvec,U,1)
plt.plot(xvec,U)
plt.show()
```
